```python
from google.colab import output
from time import sleep

i = 0

while i < 5:
    num1 = float(input('Digite o primeiro número: '))
    num2 = float(input('Digite o segundo número: '))
    i += 1
    output.clear()

    op = input("""
        [1] Soma
        [2] Substração
        [3] Multiplicação
        [4] Divisão
        [5] Sair\n
    """)

    if op == '1':
        print(num1 + num2)
    elif op == '2':
        print(num1 - num2)
    elif op == '3':
        print(num1 * num2)
    elif op == '4':
        print(num1 / num2)
    elif op == '5':
        break
    else:
        print('Operação invalida')

    sleep(5)
    output.clear()
```

```python
# Acerte a letra

import string
from random import choice
from time import sleep
from google.colab import output

alfabeto_minusculo = list(string.ascii_lowercase)

letra_secreta = choice(alfabeto_minusculo)

while True:
    letra = input('Digite uma letra: \n').lower()

    if letra == letra_secreta:
        print('Vocês acertou a letra!')
        break
    elif letra > letra_secreta:
        print('Esta letra é maior!')
    elif letra < letra_secreta:
        print('Esta letra é menor!')
    else:
        print('Entrada inválida')
    sleep(3)
    output.clear()
```

```python
# Exercicio:
# Desenvolver um programa que vai solicitar diversos números (positivos/negativos)
# e após o comando de saída ele vai exibir qual o maior número
# e qual o menor número digitado, bem como a quantidade de números digitados.
```

```python
maior_numero = 0
menor_numero = 0
quantidade_numeros = 0

print('Digite os números (ou digite "sair" para finalizar):')

while True:
    numero = input()

    if numero.lower() == 'sair':
        break

    else:
        numero = float(numero)
        quantidade_numeros += 1

        if maior_numero == 0 or numero > maior_numero:
            maior_numero = numero
        if menor_numero == 0 or numero < menor_numero:
            menor_numero = numero

if quantidade_numeros > 0:
    print(f'\nResultados:')
    print(f'Maior número digitado: {maior_numero}')
    print(f'Menor número digitado: {menor_numero}')
    print(f'Quantidade de números digitados: {quantidade_numeros}')
else:
    print('\nNenhum número foi digitado.')
```

```python
quantidade = 0
maior = float('-inf')
menor = float('inf')


while True:
    numero = input('Digite um número qualquer ("S" para sair: ').lower()


    if numero == 's':
        break
    else:
        numero = float(numero)
        quantidade += 1


        if numero > maior:
            maior = numero
        if numero < menor:
            menor = numero


print(f'\nNúmeros digitados: {quantidade}')
print(f'O maior número foi: {maior}')
print(f'O menor número foi: {menor}')
```

```python
# faça um programa que escolha um número aleatório entre 1 e 500
# tente adivinhar o número; o programa te dará dicas se é maior ou menor

numero_secreto = 42

while True:
    numero = int(input("Adivinhe o número: \n"))

    if numero > numero_secreto:
        print(f'o número {numero} é maior que o número secreto')
    elif numero < numero_secreto:
        print(f'o número {numero} é menor que o número secreto')
    else:
        print(f'Parabéns você acertou o número é {numero_secreto}')
        break
```

```python
# faça um programa que escolha um número aleatório entre 1 e 500
# tente adivinhar o número; o programa te dará dicas se é maior ou menor


import random
from time import sleep
from google.colab import output


numero_secreto = random.randint(1, 500)


while True:
    numero = int(input('Adivinhe o número (1 a 500): \n'))


    if numero == numero_secreto:
        print('Você acertou o número 🥳🎉!')
        break
    elif numero > numero_secreto:
        print('Seu número é maior! ⬆️')
    elif numero < numero_secreto:
        print('Seu número é menor! ⬇️')
    else:
        print('Entrada inválida')
    sleep(3)
    output.clear()
```
