```python
# Tratamento de Dados

# Forçar a conversão das colunas para números (o que for erro vira NaN)

colunas_numericas = ['Quantidade','Preço Unitário', 'Total Gasto']
for col in colunas_numericas:
    df_cafeteria[col] = pd.to_numeric(df_cafeteria[col], errors='coerce')
```

```python
df_cafeteria.info()
```

```python
# Aplicar a regra de três vias para preencher os buracos (NaN)

df_cafeteria['Total Gasto'] = df_cafeteria['Total Gasto'].fillna(df_cafeteria['Quantidade'] * df_cafeteria['Preço Unitário'])
df_cafeteria['Preço Unitário'] = df_cafeteria['Preço Unitário'].fillna(df_cafeteria['Total Gasto'] / df_cafeteria['Quantidade'])
df_cafeteria['Quantidade'] = df_cafeteria['Quantidade'].fillna(df_cafeteria['Total Gasto'] / df_cafeteria['Preço Unitário'])
```

```python
df_cafeteria.info()
```

```python
df_cafeteria[colunas_numericas].isnull().sum()
```

```python
# Filtrar as linhas que possuem algum valor vazio (NaN) nas colunas númericas

linhas_vazias = df_cafeteria[df_cafeteria[colunas_numericas].isnull().any(axis=1)]
display(linhas_vazias)
```
