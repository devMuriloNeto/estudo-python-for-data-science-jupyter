```python
# Aprovação de Empréstimo

# Objetivo: Verificar aprovação de empréstimo com novos critérios:

# Valor do empréstimo (R$)
# Prazo (meses)
# Renda

# Cálculo Automático: Valor da parcela = (valor do empréstimo / prazo)

# Requisitos para aprovação:

# Renda mínima: R$ 3.000
# Score de crédito: >= 500
# Parcela máxima: <= 35% da renda

renda = float(input('Informe a renda: '))
score = int(input('Informe o Score: '))
prazo = 12

if renda >= 3000 and score >= 500:
    val_emprestimo = float(input('Valor do emprestimo: '))
    if (val_emprestimo / prazo) <= ((renda / 100) * 35):
        print(f'Emprestimo de R$ {val_emprestimo:.2f} aprovado para pagamento em {prazo} meses')
    else:
        print('Emprestimo reprovado')
else:
    print('Emprestimo reprovado')
```

```python
renda = float(input('Digite o valor da sua renda: '))
score = float(input('Digite o valor do score informado pelo SERASA: '))


if renda >= 3000 and score >= 500:
    renda35 = renda * (35/100)
    print(f'Você pode comprometer até R$ {renda35:.2f} da sua renda.')

    vr_emprestimo = float(input('Informe o valor do empréstimo: '))
    qt_parcelas = int(input('Informe a quantidade de parcelas: (Máx:36) '))

    if qt_parcelas <= 36:
        vr_parcela = vr_emprestimo / qt_parcelas

        if vr_parcela <= renda35:
            print(f'Empréstimo aprovado em {qt_parcelas}x de R$ {vr_parcela:.2f} ')
        else:
            print(f'Valor da parcela maior do que o valor permitido! ')
            print(f'O valor da parcela escolhida foi R$ {vr_parcela:.2f}')
    else:
        print('Número de parcelas superior ao permitido! ')
else:
    print('Empréstimo negado!')
```

```python
# Exercício: Cálculo de frete

# Você precisa programar a lógica de cálculo de frete para uma loja online.
# O custo do frete não é fixo; ele depende do valor da compra, do peso do
# produto e da região de entrega.

# Regras:
# Frete Grátis: Compras acima de R$ 200,00 têm frete grátis
# Custo por Região (para compras até R$ 200,00):
#        Capital:  Custo R$ 10,00.
#        Interior: Custo base de R$ 15,00.
#
# Adicional por Peso (para compras até R$ 200,00):
#        Se o pacote pesar mais de 2 kg, um adicional de R$ 5,00 é somado
#
# Frete Expresso:
#        O cliente pode optar por um frete expresso, que custa R$ 8,00 a mais.
#        Esse valor é adicionado ao custo total do frete
#       (mesmo que seja frete grátis, o expresso é cobrado)

valor = float(input('Valor: '))

if valor >= 200:
    print('VocÊ possui frete gratis')
elif valor < 200:
    peso = float(input('Peso: '))
    if peso > 2:
        valor = valor + 5
    regiao = int(input('Informe a região de entrega: \n1 - Capital \n2 - Interior \n'))
    if regiao == 1:
        frete = 10
    else:
        frete = 15
    print(f'Frete de R$ {frete:.2f} reais')


```

```python
valor_compra = float(input('Valor da compra: '))
frete_total = 0

if valor_compra > 200:
    print('Você possui frete grátis.')
    frete = 0
else:
    regiao = int(input('Informe a região de entrega: \n1 - Capital \n2 - Interior \n'))
    if regiao == 1:
        frete = 10
    else:
        frete = 15

    peso = float(input('Peso do pacote (em kg): '))
    if peso > 2:
        print('Adicional de R$ 5,00 por peso aplicado para pedidos com mais de 2kg.')
        frete += 5

frete_total = frete

frete_exp = input('\nVocê deseja frete expresso (custo + R$ 8,00)? \n1 - Sim \n2 - Não\n')
if frete_exp == '1':
    print('Frete expresso selecionado (+R$ 8,00).')
    frete_total += 8

print(f'\nValor total do frete: R$ {frete_total:.2f}')
print(f'Valor total da compra com frete: R$ {(valor_compra + frete_total):.2f}')
```

```python
# Faça um programa que leia o Nome do aluno, Nome da Matéria,
# as Notas separadas do 1º Bimestre, 2º Bimestre, 3º Bimestre e 4º Bimestre.
# Em seguida calcule a Média.


# Apresente na Tela as seguintes informações:


# ----------------------
#     Boletim Escolar
# ----------------------
# Matéria:
# Nome do aluno:
# Média:
# Situação
# ----------------------


# Matéria: (Mostre o nome da matéria)
# Nome do aluno : (Mostre o nome do aluno)
# Média Calculada: (Mostre a média e se quiser as notas)
# Situação: (Mostre Aprovado, Recuperação ou Reprovado)
# Média >= 8 --> aprovado   >= 5 --> recuperação  < 5 -- > reprovado

nome = input('Digite seu nome: ')
materia = input('Nome da materia: ')
nota1 = float(input("Digite a sua nota do 1º Bimestre: "))
nota2 = float(input("Digite a sua nota do 2º Bimestre: "))
nota3 = float(input("Digite a sua nota do 3º Bimestre: "))
nota4 = float(input("Digite a sua nota do 4º Bimestre: "))
media = (nota1 + nota2 + nota3 + nota4) / 4
situacao = 'NA'
if media >= 8:
    situacao = 'aprovado'
elif media >= 5:
    situacao = 'recuperação'
else:
    situacao = 'reprovado'


print(40 * '-')
print(f'{"Boletim Escolar":^40}')
print(40 * '-')
print(f'{'Materia:':<20}', end='')
print(f'{materia:>20}')
print(f'{'Nome:':<20}', end='')
print(f'{nome:>20}')
print(f'{'Média Calculada:':<20}', end='')
print(f'{media:>20}')
print(f'{'Situação:':<20}', end='')
print(f'{situacao:>20}')
print(40 * '-')
```

```python
from random import choice

opcao = ['Pedra','Papel','Tesoura']
atom = choice(opcao)

print(f'{'\nEscolha entre: '}\n1 - Pedra \U0000270a \n2 - Papel \U0000270b \n3 - Tesoura \U0000270c: ')

while True:
    humano_input = input()
    if humano_input in ['1', '2', '3']:
        break
    else:
        print('Escolha inválida. Por favor, digite 1, 2 ou 3.')

humano_escolha = opcao[int(humano_input) - 1]

print(f'Você escolheu: {humano_escolha}')
print(f'O computador escolheu: {atom}')

if humano_escolha == atom:
    print('Empate!')
elif (
    (humano_escolha == 'Pedra' and atom == 'Tesoura') or
    (humano_escolha == 'Papel' and atom == 'Pedra') or
    (humano_escolha == 'Tesoura' and atom == 'Papel')
):
    print('Você ganhou!')
else:
    print('O computador ganhou!')
```
