## Passo 3: Análise Descritiva e Perguntas de Negócio
7. Qual foi o faturamento analisado?
8. Qual loja teve o maior faturamento? Grafico com as 5 primeiras lojas de barra.
9. Qual produto foi o mais vendido em quantidade? Grafico com os 3 primeiros barra.
10. Qual produto foi o menos vendido em quantidade?
11. Qual produto gerou a maior receita?
12. Qual foi o lucro total e a margem de lucro geral?

```python
df_final.drop_duplicates(inplace=True)
df_final.duplicated().sum()
```

```python
df_final['valor_final'].sum()
faturamento = df_final['valor_final'].sum()

faturamento
```

```python
df_final['lucro'] = (df_final['valor_unitario'] - df_final['valor_custo']) * df_final['quantidade']
df_final
```

```python
df_final['margem_lucro'] = (((df_final['valor_unitario'] - df_final['valor_custo']) / df_final['valor_unitario']) * 100).round(2)
df_final
```

```python
# 7) Qual foi o faturamento analisado?

faturamento_lojas = df_final.groupby('id_loja')[['valor_final']].sum().reset_index()
faturamento_lojas.head()
```

```python
faturamento_lojas = df_final.groupby('id_loja')['valor_final'].sum().sort_values(ascending=False).reset_index()
faturamento_lojas.head()

faturamento_lojas_top5 = faturamento_lojas.head()

display(faturamento_lojas_top5)
```

```python
# 8) Qual loja teve o maior faturamento? Grafico com as 5 primeiras lojas de barra.

fig = px.bar(faturamento_lojas_top5, x='id_loja', y='valor_final', title='Faturamento por Loja', text=faturamento_lojas_top5['valor_final'], color='valor_final', color_continuous_scale=px.colors.qualitative.Set1)

fig.show()
```

```python
# 9) Qual produto foi o mais vendido em quantidade? Grafico com os 3 primeiros barra.

produto_mais = df_final.groupby('produto')['quantidade'].sum().sort_values(ascending=False).reset_index()

display(produto_mais)
```

```python
top_produtos = df_final.groupby('produto')['quantidade'].sum().sort_values(ascending=False).reset_index()
top_produtos.head(3)
fig = px.bar(top_produtos.head(3), x='produto', y='quantidade', text=top_produtos['quantidade'].head(3), color='produto')
fig.update_traces(textposition='outside')
fig.show()
```

```python
# 10) Qual produto foi o menos vendido em quantidade?

menos_vendido = top_produtos['produto'].tail(1)
print(f'Produto menos vendido: {menos_vendido}')
```

```python
# Qual foi o lucro total e a margem de lucro geral?

receita = df_final['lucro'].sum()
print(f'Receita Total: {receita:,.2f}')

margem_geral = receita / faturamento * 100
print(f'Lucro Geral: {margem_geral:.2f} %')
```

## Passo 4: Gerentes
13. Qual foi o mês com maior faturamento? Grafico de linha
14. Quais foram os produtos mais vendidos por cada gerente? Grafico de arvore

```python
# Qual foi o mês com maior faturamento? Grafico de linha

df_final['ano - mes'] = df_final['data'].dt.to_period('M')
df_final
```

```python
df_final['ano - mes'] = df_final['data'].dt.to_period('M')
fat_mes = df_final.groupby('ano - mes')['valor_final'].sum().reset_index()

fat_mes['ano - mes'] = fat_mes['ano - mes'].dt.to_timestamp()

fig = px.line(fat_mes, x='ano - mes', y='valor_final', line_shape='spline', markers='1')
fig.show()
```

```python
# Quais foram os produtos mais vendidos por cada gerente? Grafico de arvore

top_pro_gerentes = df_final.groupby(['gerente','produto'])['quantidade'].sum().reset_index().sort_values(by='quantidade')

fig = px.treemap(top_pro_gerentes, path=[px.Constant('Gerentes'),'gerente','produto'], values='quantidade')
fig.show()
```
